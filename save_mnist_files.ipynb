```python
import tensorflow as tf
import tensorflow_datasets as tfds
import os

# Load the MNIST dataset using TensorFlow Datasets
mnist_dataset, mnist_info = tfds.load(name='mnist', with_info=True, as_supervised=True)

# Get the training dataset
train_dataset = mnist_dataset['test']

# Create a directory to save the images
save_dir = "mnist_test_images"
for i in range(10):
    directory_name = str(i)
    os.makedirs(save_dir+'/'+directory_name, exist_ok=True)
```

```python

# Save the images with labels as file names
for idx,example in enumerate(train_dataset):
    image, label = example
    image = tf.cast(image, tf.uint8)
    image = tf.image.encode_png(image)
    image_label = str(label.numpy())
    filename = os.path.join(save_dir + '/' + image_label, f'{idx}{label.numpy()}.jpg')
    tf.io.write_file(filename, image)
    print(filename)
```
